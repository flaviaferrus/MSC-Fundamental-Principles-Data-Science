Let us compute the Newton method for the initial points considered inside both intervals. I may consider for example the middle point of the interval in order to do so, and compute the gradient and the Hessian as aforementioned analytically (remember that in this case the function goes from the reals to the reals (it is one dimensional problem).

```python
import numpy as np

def f(x):
    return x**2 + np.exp(x) - 3

def f1(x):
    return 2*x + np.exp(x)

def f2(x):
    return 2 + np.exp(x)

import matplotlib.pyplot as plt
x=np.arange(-2,2, 0.1)
y = f(x)

plt.plot(x, y)
plt.axis(True)
```

```python
def Newton(x0, thres=1e-3, alpha=0.1, maxSteps=100):
    x_path=np.zeros(maxSteps)
    x_path[0]=x0
    x_0 = x0
    #alpha=0.5
    for i in range(1,100):
        d_k=f(x_0)/f1(x_0)
        x_k=x_0-alpha*d_k
        #if f(x_k)<f(x_0):
        if(abs(f(x_0))) < thres:
        #if abs(f(x_k) - f(x_0)) < thres:
            return x_path[0:i], i
        else:
            x_0 = x_k
        x_path[i]=x_k

    return x_path,i
```

```python
x1 = -1.5
x2 = 1

xn1, it1 = Newton(x1)
xn2, it2 = Newton(x2)

plt.plot(x, y)
plt.plot(xn1, f(xn1), 'r')
plt.plot(xn2, f(xn2), 'g')
```

```python
import numpy as np

def f(x):
    return x**2 + np.exp(x) - 3

def f1(x):
    return 2*x + np.exp(x)

def f2(x):
    return 2 + np.exp(x)

import matplotlib.pyplot as plt
x=np.arange(-2,2, 0.1)
y = f(x)

plt.plot(x, y)
plt.axis(True)

def my_newton(f, df, x0, tol):
    # output is an estimation of the root of f 
    # using the Newton Raphson method
    # recursive implementation
    if abs(f(x0)) < tol:
        return x0
    else:
        return my_newton(f, df, x0 - f(x0)/df(x0), tol)
```

```python
estimate = my_newton(f, f2, -1.25, 1e-6)
print("estimate =", estimate)
```
